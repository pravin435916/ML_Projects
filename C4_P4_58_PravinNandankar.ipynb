## 1. Dataset Explanation
The Titanic dataset usually has the following attributes:
- PassengerId → unique ID for each passenger
- Survived → target variable (0 = No, 1 = Yes)
- Pclass → Ticket class (1st, 2nd, 3rd)
- Sex → Male/Female
- Age → Age of the passenger
- SibSp → Number of siblings/spouses aboard
- Parch → Number of parents/children aboard
- Fare → Ticket fare
- Embarked → Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
# load dataset
file_path = "/content/Titanic-Dataset.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df.info()
```

```python
np.random.seed(42)  #for reproducibility
df['Sex'] = np.random.choice(['male', 'female'], size=len(df))
df['Age'] = np.random.randint(1, 80, size=len(df))
```

```python
contingency_table = pd.crosstab(df['Sex'], df['Survived'])
contingency_table
```

## 2. Chi-Square Test
We perform the Chi-Square Test of Independence between:
● Sex (Male/Female)
● Survived (0 = Did not survive, 1 = Survived)

This helps check if survival on the Titanic depended on gender.

```python
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table: \n{expected}")
```

```python
# Give a description about insight drawn after performing the test .
```

```python
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis. There is a significant difference in survival rates between male and female passengers.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in survival rates between male and female passengers.")
```

## Chi-Square Test
We tested the relationship between gender and survival status on the Titanic.
The result showed that the p-value was greater than 0.05, which means there is no significant
association between gender and survival in our simulated dataset.
In other words, male and female passengers were statistically equally likely to survive or not
survive.
Thus, gender does not play a significant role in determining survival in this dataset.

## 3.Paired T-Test

- Before Scores (simulated passenger performance before an intervention)
- After Scores (simulated passenger performance after an intervention)

This checks if there is a significant difference between the two sets of scores (before vs.
after).

### Parameters/Attributes on which Paired t-Test Test is performed

```python
from scipy.stats import ttest_rel
np.random.seed(42)
df['Before Scores'] = np.random.normal(70, 10, size=len(df)) # mean=70,std=10
df['After Scores'] = df['Before Scores'] + np.random.normal(0, 2, size=len(df)) # add some random noise
```

```python
t_stat , p_value = ttest_rel(df['Before Scores'], df['After Scores'])
print(f"T-Statistic: {t_stat}")
print(f"P-value: {p_value}")
```

## Insight:
- The After Scores are statistically higher than the Before Scores, indicating the
- intervention (training/condition) had a positive impact.

## Result of Paired t-Test (Simulated Titanic Data)
### We tested the difference between Before Scores and After Scores (simulated paired data for
Titanic passengers).
- The Paired t-Test gave a p-value less than 0.05.
- This means we reject the null hypothesis and conclude that there is a significant
## difference between Before and After scores.

### In other words, the “After” scores are statistically higher than the “Before” scores, which
### indicates that the intervention (training/condition) had a positive impact.
